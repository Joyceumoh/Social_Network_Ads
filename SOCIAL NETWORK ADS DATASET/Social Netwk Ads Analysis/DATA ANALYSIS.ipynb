# UNIVARIATE EXPLORATION

```python
plt.figure(figsize=(10, 5))
sns.histplot(data['Age'], bins=5)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Persons from 35 to 42 years old have the highest frequency in the dataset
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data['EstimatedSalary'], bins=5)
plt.title('Estimated Salary Distribution')
plt.xlabel('Estimated Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Estimated salary from 70,000 to 95,000 have the highest frequency in the dataset
```

```python
# Histograms to visualize distributions
plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.hist(data['Purchased'], bins=2, color='purple', alpha=1)
plt.title('Purchased Distribution')
plt.xlabel('Purchased')
```

```python
# People who did not purchase are more than those who purchase
```

# BIVARIATE EXPLORATION

```python
sns.scatterplot(x='Age', y='Purchased', data=data)
plt.title('Purchased vs Age')
plt.show()
```

```python
# There is correlation between purchased and the ages of people who purchased
```

```python
sns.scatterplot(x='EstimatedSalary', y='Purchased', data=data)
plt.title('Purchased vs EstimatedSalary')
plt.show()
```

```python
# There's relationship between those who puchased and their estimated salaries
```

```python
sns.scatterplot(x='Age', y='EstimatedSalary', data=data)
plt.title('EstimatedSalary vs Age')
plt.show()
```

```python
sns.boxplot(x='Purchased', y='Age', data=data)
plt.title('Age vs. Purchase')
plt.show()
```

```python
# Persons from 17 to 38 years old did not make a purchase
# Persons from 41 to 60 years old made a purchase
```

```python
sns.boxplot(x='Purchased', y='EstimatedSalary', data=data)
plt.title('Estimated Salary vs. Purchased')
plt.show()
```

```python
# Persons earning 15,000 to 77,000 did not make a purchase
# Persons earning 41,000 to 150,000 made a purchase
```

```python
sns.kdeplot(data[data['Purchased'] == 1]['EstimatedSalary'], label='Purchased')
sns.kdeplot(data[data['Purchased'] == 0]['EstimatedSalary'], label='Not Purchased')
plt.title('Estimated Salary Distribution')
plt.xlabel('Estimated Salary')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Persons who made a purchase earn more salary than those who did not make a purchase.
```

```python
sns.boxplot(data=data[['Age', 'EstimatedSalary']])
plt.title('Boxplot of Age and Estimated Salary')
plt.show()
```

```python
# There are no outliers in Age and EstimatedSalary columns and this means that no value in the columns are above the mean values
# It also means that there is no bias 
```

# Multivariate Exploration

```python
corr_matrix=data.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='viridis',)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Age column has more relationship (62%) with Purchased column compared to EstimatedSalary (36%)
```

```python
# Pairplot to visualize relationships between all variables
sns.pairplot(data)
```

```python
# The above pairplot shows that Age has a relationship with Purchased
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['Age'], data['EstimatedSalary'], data['Purchased'], c='r', marker=0)
ax.set_xlabel('Age')
ax.set_ylabel('EstimatedSalary')
ax.set_zlabel('Purchased')
plt.title('3D Scatter Plot of Age, EstimatedSalary and Purchased')
```

```python
# This shows that there's a relationship between Age, EstimatedSalary and those who purchased.
```
